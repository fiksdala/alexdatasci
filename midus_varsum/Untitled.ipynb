```python
def getdict(varname):
    keys=None
    values=None
    for i in root.findall('dataDscr/var'):
        if i.attrib['name'] == varname:
            keys = i.findall('catgry/labl')
            values = i.findall('catgry/catValu')
    if keys is not None:
        output=dict(zip([i.text for i in keys],[i.text for i in values]))
    else:
        output = dict()
    return output;
```
